# **●AIエージェント構想案**

このAIエージェントの構想は、単なる「AI関連銘柄探し」ではなく、

> **企業のAI活用度（AI Engagement）を定量化し、その将来キャッシュフロー改善可能性を先行的に評価することで、市場がまだ十分に織り込んでいないAIプレミアムを発見する**

ことを目的としています。

特に理論的なベースとしては、

* 「企業のAI関与度（AI Engagement）をテキスト分析で定量化できる」
* 「AI関与度の高い企業ほど、ChatGPT登場後に有意な異常リターン（abnormal return）を示した」
* 「既存のAI ETFよりも、NLPによるAI Engagement Indexの方が透明性・説明可能性が高い」

ことを示した論文

**"Quantifying A Firm's AI Engagement: Constructing Objective, Data-Driven, AI Stock Indices Using 10-K Filings"**（Ante & Saggu, 2025）を直接の理論的出発点とする。([arXiv][1])

### また、[こちらのYilei Shao論文](https://journals.sagepub.com/doi/full/10.1177/20965311261422769)にあるSilicon-Based Economyというコンセプトもベースにしている。

````python
schema = {
    "type": "json_schema",
    "name": "ai_adoption_score",
    "schema": {
        "type": "object",
        "properties": {
            "ai_adoption": {"type": "integer", "minimum": 0, "maximum": 100},
            "agent_readiness": {"type": "integer", "minimum": 0, "maximum": 100},
            "labor_replacement_potential": {"type": "integer", "minimum": 0, "maximum": 100},
            "ai_business_impact": {"type": "integer", "minimum": 0, "maximum": 100},
            "reason": {"type": "string"}
        },
        "required": [
            "ai_adoption",
            "agent_readiness",
            "labor_replacement_potential",
            "ai_business_impact",
            "reason"
        ],
        "additionalProperties": False
    },
    "strict": True
}
````

````python
def score_company_with_llm(company):
    prompt = f"""
あなたは日本株のAI活用度を評価する投資分析エージェントです。

企業名: {company["name"]}
証券コード: {company["code"]}

IR・開示テキスト:
{company["ir_text"]}

以下を0〜100で採点してください。

ai_adoption:
AIをどれほど本格的に導入しているか。

agent_readiness:
AIエージェントに業務を任せられる構造があるか。

labor_replacement_potential:
AIによって人件費・外注費・ホワイトカラー業務を代替できる余地。

ai_business_impact:
AI活用が売上、利益率、競争力に直結しそうか。

投資推奨ではなく、あくまでAI活用度評価として採点してください。
"""

    response = client.responses.create(
        model="gpt-5.5",
        input=prompt,
        text={"format": schema}
    )

    return json.loads(response.output_text)
````

````python
rows = []

for company in companies:
    llm_score = score_company_with_llm(company)
    stock = get_stock_features(company["code"])

    ai_alpha_score = (
        0.30 * llm_score["ai_adoption"]
        + 0.25 * llm_score["agent_readiness"]
        + 0.25 * llm_score["labor_replacement_potential"]
        + 0.20 * llm_score["ai_business_impact"]
    )

    rows.append({
        "code": company["code"],
        "name": company["name"],
        **llm_score,
        **stock,
        "ai_alpha_score": round(ai_alpha_score, 2)
    })

df = pd.DataFrame(rows)
df.sort_values("ai_alpha_score", ascending=False)
````
